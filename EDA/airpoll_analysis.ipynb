```python
df['Avg']=df['Avg'].fillna(df['Avg'].mean())
```

```python
df['Min']=df['Min'].fillna(df['Min'].mean())
```

```python
df['Max']=df['Max'].fillna(df['Max'].mean())
```

```python
df.isnull().sum()
```

```python
df.drop(['Country','lastupdate'],axis=1,inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(17,7),dpi=100)
sns.countplot(df['State'],data=df)
plt.xlabel('State')
plt.ylabel('Pollution Level')
plt.xticks(rotation=75)
plt.show()
```

```python
df_state=df.groupby('State')
```

```python
df_state.head()
```

```python
df_state.mean()
```

```python
plt.figure(figsize=(17,7))
plt.plot(df_state.mean())
plt.tight_layout()
plt.xlabel('State')
plt.ylabel('Pollutin level')
plt.legend(['Avg','Max','Min'])
plt.xticks(rotation=75)
plt.show()
```

```python
print('Cities having max pollution wrt state')
df_state.max()['city']
```

```python
print('Cities having min pollution wrt state')
df_state.min()['city']
```

```python
import matplotlib.pyplot as plt
state=list(df['State'].unique())
f,axes=plt.subplots(5,4,figsize=(17,20))
i=0
for a in state:
    airq=df[df['State']==a]
    sns.countplot('Pollutants',data=airq,ax=axes.flatten()[i])
    plt.title(a)
    plt.tight_layout()
    i=i+1
    
```
